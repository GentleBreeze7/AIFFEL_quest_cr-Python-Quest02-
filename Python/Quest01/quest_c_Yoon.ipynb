```python
'''
Quest 1. 회문 판별기

Guide : https://wikidocs.net/7896

A. Requirements Getting Started:

https://docs.google.com/presentation/d/1Ht4F9frt8ZUwPpv_Ky-qdZLfPjP0mdSi/edit#slide=id.p5

A-1. Requirements:
1. 단어를 입력받을 수 있게 한다.
2. 입력된 단어를 뒤집어서 출력할 수 있게 한다.
3. 입력된 단어가 회문이 맞는지 확인하여 출력할 수 있게 한다.
4. 자료형, 제어문, 함수를 활용한다.

A-2. Input & Output Sample :
입력값 :
우성우

출력값 :
뒤집힌 단어는: 우성우
입력된 단어는 회문입니다.

A-3. Need more information?
- 구, 문장, 숫자 및 문단 의 입력값 예시가 필요합니다.
- 회문이 아닌 경우에 대한 출력값 예시가 필요합니다.
- 입력값의 길이에 대한 제한이 있나요?
- 대소문자를 구분하는지 확인이 필요합니다.
- 입력값이 없을 경우에 대한 처리가 필요합니다.
  - 예) 입력값이 없을 경우, 다시 입력하라는 메시지를 출력합니다.

B. Requirements Analysis:
: 오직 회문에 관련된 작은 파트의 기능만 존재.
제품 : 회문 판별기
 - 기능 #1 :  입력받기
    - input() 함수를 사용하여 사용자로부터 단어를 입력받습니다.
 - 기능 #1-1 (Optional) : 유효성 검사 (Validation) 추가적인 제약사항이 발생했을 경우
    예) 입력값이 없을 경우, 다시 입력하라는 메시지를 출력합니다.
 - 기능 #2 : 입력 받은 값 역순으로 저장하기
    - for, slicing, string -> list 변환 후 reverse() 함수 사용
 - 기능 #3 : 역순으로 저장한 값 출력하기
    - print() 함수를 사용하여 역순으로 저장한 값을 출력합니다.
 - 기능 #4 : 회문 판별하기
    - 회문 판별을 위해 입력받은 값과 역순으로 저장한 값을 비교합니다.
    - 회문이 맞는지 아닌지 출력합니다.

C. Implementations (Pair Programming):
C-1. Code Review:
D. Testing:
E. Launching:
F. Post-Mortem:
E. kudos(Optional):

'''
information_message = {"input": "단어를 입력하세요:",
                       "output": "뒤집힌 단어는:",
                       "palindrome": "입력된 단어는 회문입니다.",
                       "not_palindrome": "입력된 단어는 회문이 아닙니다."}

# 함수 : 사용자로부터 입력
def input_from_user():
    return str(input(information_message["input"]))

# 함수 : 문자열 역순으로 재배치
# Case 1: Using slicing
# reversed_word = word[::-1]
# -> 문자열에서의 슬라이스 관련한 부분에 대해 문법 정리 필요
# Case 2: Using for loop
# for i in word:
#     reversed_word = i + reversed_word
# Case 3: Using reversed() function
# reversed_word = ''.join(reversed(word))
# Case 4: Using list
# reversed_word = list(word).reverse()
# 1 방법으로 진행하는 건 구현상 코드가 적게 들어가긴 하지만 해당 구문이 이해되지 않는 경우가 많다 ( 나도 바로 보고 이해 안됨)
# [::] 입력되는 조건은 3개가 있는 것 같은데 첫번재는 시작이라고 알고 있고 , 두번째 조건은 마지막 그리고 , 세번째 조건이 증가값인데 마이너스로 적으면 반대로 작동하나봄
def reverse_string(word):
    return ''.join(reversed(word))

# 함수 : 회문 판별
def check_to_palindrome(word, reversed_word):
    # 회문 판별하기
    if word == reversed_word:
        print(information_message["palindrome"])
    else:
        print(information_message["not_palindrome"])

# 입력받기
word = input_from_user()

# 입력받은 값 역순으로 저장하기
reversed_word = reverse_string(word)

# 역순으로 저장한 값 출력하기
print(f"{information_message["output"]} {reversed_word}")

# 회문 판별하기
check_to_palindrome(word, reversed_word)
```
